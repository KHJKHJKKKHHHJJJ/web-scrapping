```python
def scr_name(url):
  response = requests.get(url)
  soup = BeautifulSoup(response.text, "html.parser")
  print(soup.find('a', {'class' : 'css-19v1rkv', 'role' : 'link'}))
  return [i.text for i in soup.find_all('a',{'class' : 'css-19v1rkv', 'role' : 'link'})]
```

```python
namelst = []
for i in range(1, 25):
  print(scr_name(f"https://www.yelp.com/search?find_desc=Restaurants&find_loc=San+Antonio%2C+TX+78205&start={i*10 - 10}"))
  namelst.extend(scr_name(f"https://www.yelp.com/search?find_desc=Restaurants&find_loc=San+Antonio%2C+TX+78205&start={i*10 - 10}"))
namelst
```

```python
ratelst = []
for i in range(1, 25):
    url = f"https://www.yelp.com/search?find_desc=Restaurants&find_loc=San+Antonio%2C+TX+78205&start={i*10 - 10}"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    ratings = soup.find_all('div', class_='css-14g69b3')  #['aria-label'])
    for rating in ratings:
        print(rating['aria-label'])
        ratelst.append(rating['aria-label'])
```

```python
def get_categ(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    all_product_element_list = soup.find_all('div', class_='priceCategory__09f24___4Wsg iaPriceCategory__09f24__x9YrM css-1941ymj')

    items_cat = []
    for item_soup in all_product_element_list:
        item_cat = [item.text for item in BeautifulSoup(str(item_soup), "html.parser").find_all('a', {'location' : '[object Object]', 'class': 'css-abnp9g'})]
        items_cat.append(item_cat)
    return items_cat
    
categlst = []
for i in range(1, 25):
  print(get_categ(f"https://www.yelp.com/search?find_desc=Restaurants&find_loc=San+Antonio%2C+TX+78205&start={i*10 - 10}"))
  categlst.extend(get_categ(f"https://www.yelp.com/search?find_desc=Restaurants&find_loc=San+Antonio%2C+TX+78205&start={i*10 - 10}"))
```

```python
import pandas as pd

res_dict = {
    'name' : namelst,
    'rate' : ratelst,
    'categories': categlst
}

res_df = pd.DataFrame(res_dict)
res_df
```

```python
def get_url(url):
    # url = "https://www.yelp.com/search?find_desc=Restaurants&find_loc=San+Antonio%2C+TX+78205"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    urlist = []
    urlsoup = soup.find_all('a', {'class' :'css-19v1rkv', 'rel': 'noopener'})

    for url in urlsoup:
        print(url['href'])
        urlist.append(url['href'])
    return urlist

urlist = []
for i in range(1, 25):
  print(get_url(f"https://www.yelp.com/search?find_desc=Restaurants&find_loc=San+Antonio%2C+TX+78205&start={i*10 - 10}"))
  urlist.extend(get_url(f"https://www.yelp.com/search?find_desc=Restaurants&find_loc=San+Antonio%2C+TX+78205&start={i*10 - 10}"))
```

```python
import requests
from bs4 import BeautifulSoup

addrlist = []
for urls in urlist: 
    tried = 1 
    while tried <= 5:
        try:
            url = "https://www.yelp.com" + urls
            response = requests.get(url)
            soup = BeautifulSoup(response.text, "html.parser")
            address = soup.find('p', {'class':'css-qyp8bo'})
            # if address:
            print(namelst[urlist.index(urls)])
            addrlist.append(address.text)
        except:
            if tried <= 5:
                print("finding again...")
                tried += 1
            else:
                print("we couldn't find it.")
                addrlist.append('None')
                break
        break
addrlist
```
